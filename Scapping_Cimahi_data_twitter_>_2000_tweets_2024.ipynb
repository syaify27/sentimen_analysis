```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Download NLTK data jika belum ada
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

# Stopword tambahan
stopword_tambahan = ['wiiir', 'yang', 'di', 'dan', 'dengan','ga','aja','yg']  # Contoh stopword tambahan

def clean_text(text):
    # Ubah teks ke lowercase
    text = text.lower()
    # Hapus karakter non-alfanumerik dan spasi berlebih
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)
    text = re.sub(r"\s+", " ", text)

    # Tokenisasi teks
    tokens = word_tokenize(text)

    # Gabungkan stopword dari NLTK dan stopword tambahan
    stop_words = set(stopwords.words('indonesian'))
    stop_words.update(stopword_tambahan)

    # Hapus stopword dari tokens
    tokens = [token for token in tokens if token not in stop_words]

    return " ".join(tokens)

# Bersihkan teks dalam variabel 'data'
data_cleaned = data.apply(clean_text)

# Gabungkan semua teks yang sudah dibersihkan
all_text = " ".join(data_cleaned.astype(str).tolist())

# Buat wordcloud
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap="YlGnBu_r").generate(all_text)

# Tampilkan wordcloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Path ke gambar PNG di Google Drive
mask_path = '/content/drive/MyDrive/peta cimahi.jpg'  # Ganti dengan path gambar Anda

# Download NLTK data jika belum ada
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

# Stopword tambahan
stopword_tambahan = ['wiiir', 'yang', 'di', 'dan', 'dengan']

def clean_text(text):
    # Ubah teks ke lowercase
    text = text.lower()
    # Hapus karakter non-alfanumerik dan spasi berlebih
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)
    text = re.sub(r"\s+", " ", text)

    # Tokenisasi teks
    tokens = word_tokenize(text)

    # Gabungkan stopword dari NLTK dan stopword tambahan
    stop_words = set(stopwords.words('indonesian'))
    stop_words.update(stopword_tambahan)

    # Hapus stopword dari tokens
    tokens = [token for token in tokens if token not in stop_words]

    return " ".join(tokens)

# Bersihkan teks dalam variabel 'data'
data_cleaned = data.apply(clean_text)

# Gabungkan semua teks yang sudah dibersihkan
all_text = " ".join(data_cleaned.astype(str).tolist())

# Baca gambar mask
mask = np.array(Image.open(mask_path))

# Buat wordcloud dengan mask
wordcloud = WordCloud(width=1600, height=800,
                      background_color='brown',
                      #colormap="YlOrRd_r",
                      mask=mask).generate(all_text)

# Buat pewarnaan wordcloud berdasarkan gambar mask (opsional)
image_colors = ImageColorGenerator(mask)

# Tampilkan wordcloud
plt.figure(figsize=(30, 15))
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="nearest") # Gunakan recolor jika ingin pewarnaan dari gambar
plt.axis("off")
plt.savefig(driveroot+fName+'1'+".png")
plt.show()
```
